## Create function to scrape the data

```python
player_stats = []
def scrape_player_by_team(team_urls):
    browser.get(team_urls)
    data = browser.page_source
    soup = BeautifulSoup(data, 'html.parser')

    standings_table = soup.select('table.stats_table')[9]

    links_players = standings_table.find_all('a')
    links_players = [l.get("href") for l in links_players]
    links_players = [l for l in links_players if l and '/matchlogs/2022-2023/' in l]
    summary_links_players = []
    for link in links_players:
        summary_links_players.append(link.replace('possession', 'summary'))

    player_urls= [f"https://fbref.com{l}" for l in summary_links_players]
    for i in player_urls:
        browser.get(i)
        data = browser.page_source
        player_stat = pd.read_html(data, match="2022-2023",header=1)[0]

        new_player_stat = player_stat[['Date','Squad','Venue','Start', 'Pos', 'Min', 'Gls', 'Ast', 'PK', 'PKatt', 'Sh', 'SoT', 'CrdY',
            'CrdR', 'Touches', 'Tkl', 'Int', 'Blocks', 'xG', 'npxG', 'xAG', 'SCA',
            'GCA', 'Cmp', 'Att', 'Cmp%', 'PrgP', 'Carries', 'PrgC', 'Att.1', 'Succ']]
        new_player_stat['Name'] = i.split('/')[-1].split('-Match-Logs')[0].replace('-', ' ')
        new_player_stat = new_player_stat.dropna()
        player_stats.append(new_player_stat)

        
        if i == player_urls[-1]:
            print('end team,', team_urls.split("/")[-1], ' ', player_stats[0])
            df = pd.concat(player_stats, axis=0, ignore_index=True)
            df.to_csv(f'./data/{team_urls.split("/")[-1]}-players-stat'+'.csv', index=False)
```

```python
for team in team_urls:
    scrape_player_by_team(team)
```

```python
for range_team in range(12,21):
    scrape_player_by_team(team_urls[range_team])
```
