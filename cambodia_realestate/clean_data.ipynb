```python
import os
import json
import re
from datetime import date
```

## Load Data

```python
def load_data(title):
    with open(title,encoding='utf-8') as f:
        return json.load(f)
```

```python
realestate_kh = "./data/realestate_kh_50_pages_2023-05-21.json"
realestate_kh_list = load_data(realestate_kh)
```

```python
realestate_kh_list
```

## Create regax to group location

```python
cambodia_province = '''Banteay Meanchey|Battambang|Kampong Cham|Kampong Chhnang|Kampong Speu|Kampong Thom|Kampot|Kandal|Koh Kong|Kratié|Kratie|Mondulkiri|Phnom Penh|Preah Vihear|Prey Veng|Pursat|Ratanakiri|Siem Reap|Preah Sihanouk|Sihanoukville|Stung Treng|Svay Rieng|Takéo|Oddar Meanchey
|Kep|Pailin|Tboung Khmum|Takeo'''
```

```python
for i in realestate_kh_list:
    if re.search(cambodia_province, str(i['location']),re.IGNORECASE):
        i['location'] = re.search(cambodia_province, str(i['location']),re.IGNORECASE).group()
    else:
        # print('========= '+str(i['location'])+' =========')
        if i['location'] == 'Krong Kracheh':
            i['location'] = 'Krong Kracheh'
        if i['location']:
            i['location'] = 'Phnom Penh'
            
realestate_kh_list
```

## Clean price field

```python
for i in realestate_kh_list:
    if re.search("save", str(i['price']),re.IGNORECASE):
        i['price'] = i['price'].split('$')[2].split('SAVE')[0]
    if re.search("permonth", str(i['price']),re.IGNORECASE):
        i['price'] = i['price'].split("permonth")[0]
    if re.search("POA", str(i['price']),re.IGNORECASE):
        i['price'] = '$0'
    if re.search("/m²", str(i['price']),re.IGNORECASE):
        i['price'] = i['price'].split('/m²')[0].replace('$','').replace(',','')
    i['price'] = (i['price'].replace('$','').replace(',',''))
    i['price'] = int(i['price'])
    print(i['price'])
```

## Save Data to visualize

```python
def save_data(title,data):
    with open(title, "w",encoding='utf-8') as outfile:
        json.dump(data, outfile, ensure_ascii=False, indent=3)
```

```python
realestate_kh = "./realestate_kh_cleaner"+str(date.today())+".json"
save_data(realestate_kh,realestate_kh_list)
```
