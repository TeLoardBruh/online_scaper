```python
import os
import json
import re
from datetime import date
import requests
import urllib.parse
import time
```

## Load Data

```python
def load_data(title):
    with open(title,encoding='utf-8') as f:
        return json.load(f)
```

```python
realestate_kh = "./data/realestate_kh_50_pages_2023-06-19.json"
realestate_kh_list = load_data(realestate_kh)
```

```python
for i in realestate_kh_list:
    i['location'] = i['location'].replace('BKK ', 'Boeng Keng Kang ').strip()

    time.sleep(3)
    print(i['location'])
    if (i['location'] == ''):
        i['location'] = 'Other'
    if (i['location'] == 'Luxcity Hotel & Apartment'):
        i['location'] = 'Other'
    if (len(i['location'].strip().split(',')) == 1):
        i['sangkat'] = None
        i['khan'] = None
        i['city/provice'] = None
        url = 'https://nominatim.openstreetmap.org/search/' + urllib.parse.quote(i['location']) +'?format=json'

        response = requests.get(url).json()
        if (response != []):
            i['lat'] = response[0]["lat"]
            i['lon'] = response[0]["lon"]
        else:
            i['lat'] = None
            i['lon'] = None
    if (len(i['location'].strip().split(',')) == 2):
        i['sangkat'] = None
        i['khan'] = i['location'].strip().split(',')[0].strip()
        i['city/provice'] = i['location'].strip().split(',')[1].strip()
        url = 'https://nominatim.openstreetmap.org/search/' + urllib.parse.quote(i['khan']) +'?format=json'

        response = requests.get(url).json()
        if (response != []):
            i['lat'] = response[0]["lat"]
            i['lon'] = response[0]["lon"]
        else:
            i['lat'] = None
            i['lon'] = None
    if (len(i['location'].strip().split(',')) == 3):
        i['sangkat'] = i['location'].strip().split(',')[0].strip()
        i['khan'] = i['location'].strip().split(',')[1].strip()
        i['city/provice'] = i['location'].strip().split(',')[2].strip()
        url = 'https://nominatim.openstreetmap.org/search/' + urllib.parse.quote(i['sangkat']) +'?format=json'

        response = requests.get(url).json()
        if (response  != []):
            i['lat'] = response[0]["lat"]
            i['lon'] = response[0]["lon"]
        else:
            i['lat'] = None
            i['lon'] = None
    if (len(i['location'].strip().split(',')) == 4):
        i['sangkat'] = i['location'].strip().split(',')[1].strip()
        i['khan'] = i['location'].strip().split(',')[2].strip()
        i['city/provice'] = i['location'].strip().split(',')[3].strip()
        url = 'https://nominatim.openstreetmap.org/search/' + urllib.parse.quote(i['sangkat']) +'?format=json'

        response = requests.get(url).json()
        if (response  != []):
            i['lat'] = response[0]["lat"]
            i['lon'] = response[0]["lon"]
        else:
            i['lat'] = None
            i['lon'] = None 

```

## Clean price field

```python
for i in realestate_kh_list:
    if re.search("save", str(i['price']),re.IGNORECASE):
        i['price'] = i['price'].split('$')[2].split('SAVE')[0]
    if re.search("permonth", str(i['price']),re.IGNORECASE):
        i['price'] = i['price'].split("permonth")[0]
    if re.search("POA", str(i['price']),re.IGNORECASE):
        i['price'] = '$0'
    if re.search("/m²", str(i['price']),re.IGNORECASE):
        i['price'] = i['price'].split('/m²')[0].replace('$','').replace(',','')
    i['price'] = (i['price'].replace('$','').replace(',',''))
    i['price'] = int(i['price'])
    print(i['price'])
```

## Save Data to visualize

```python
def save_data(title,data):
    with open(title, "w",encoding='utf-8') as outfile:
        json.dump(data, outfile, ensure_ascii=False, indent=3)
```

```python
realestate_kh = "./realestate_kh_cleaner"+str(date.today())+".json"
save_data(realestate_kh,realestate_kh_list)
```
