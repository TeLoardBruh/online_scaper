## Creating function to scrap all the product

```python
product_dictionary = []
def get_production_info(url,product_type):
    pagination_list = [url]
    browser.get(url)
    time.sleep(5)
    html = browser.page_source
    soup = bs(html, 'html.parser')
    pagination = soup.find_all(class_='woocommerce-pagination')
    print('on top')
    print(pagination_list)
    for items in pagination:
        for list_href in items.find_all('a'): 
            pagination_list.append(list_href.get('href'))
            print('in loop')
            print(pagination_list)
    pagination_list = list(dict.fromkeys(pagination_list))
    print('after convert')
    print(pagination_list)
    if len(pagination_list) > 1:
        for res in pagination_list:
            print(res)
            browser.get(res)
            time.sleep(20)
            html = browser.page_source
            soup_pagination = bs(html, 'html.parser')
            product_info =  soup_pagination.find_all(class_='mf-product-details')
            for each_product in product_info:
                product_name = each_product.find(class_="mf-product-content").text
                product_price_usd = each_product.find(class_='mf-product-price-box').find(class_="price").text
                product_price_kh = int(float(str(product_price_usd).split('$')[1]))*4060.00
                product_dictionary.append({'product_name': product_name,'typeOf': product_type, 'price_usd': product_price_usd, 'price_kh': product_price_kh})
    else:
        print(url)
        browser.get(url)
        time.sleep(20)
        html = browser.page_source
        soup_pagination = bs(html, 'html.parser')
        product_info =  soup_pagination.find_all(class_='mf-product-details')
        for each_product in product_info:
            product_name = each_product.find(class_="mf-product-content").text
            product_price_usd = each_product.find(class_='mf-product-price-box').find(class_="price").text
            product_price_kh = int(float(str(product_price_usd).split('$')[1]))*4060.00
            product_dictionary.append({'product_name': product_name,'typeOf': product_type, 'price_usd': product_price_usd, 'price_kh': product_price_kh})

```

# Saving and Loading data

```python
def save_data(title,data):
    with open(title, "w",encoding='utf-8') as outfile:
        json.dump(data, outfile, ensure_ascii=False, indent=3)
```

```python
def load_data(title):
    with open(title,encoding='utf-8') as f:
        return json.load(f)
```

```python
chipmok_store_pirce = "./data/chipmok_store_pirce.json"
save_data(chipmok_store_pirce,product_dictionary)
chipmok_store_pirce_list = load_data(chipmok_store_pirce)
```
