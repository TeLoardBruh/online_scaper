```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import re
```

# Load Data

```python
fifa = pd.read_csv('./data/players_22.csv',low_memory=False)
pd.set_option('display.max_colwidth', None)

fifa
```

```python
bins = [50,60,70,80,90,100]
plt.hist(fifa['overall'], bins=bins, color='#abcdef')
plt.xticks(bins)
plt.ylabel('Number of players')
plt.xlabel('Overall')
plt.show()
```

# Pie Charts

```python
fifa.columns
for fifa_cols in fifa:
    print(fifa_cols)
```

```python
total_left_foot = fifa.loc[fifa['preferred_foot']== 'Left'].count()[0]
total_right_foot = fifa.loc[fifa['preferred_foot']== 'Right'].count()[0]

labels_foot = ['Left','Right']
colors_labels = ['#D1FF33','#82A01B']

plt.pie([total_left_foot,total_right_foot], labels = labels_foot, colors=colors_labels, autopct='%.2f %%')
plt.title('FIFA players preferred foot')
plt.show()
```
