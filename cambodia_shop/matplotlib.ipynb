```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(8,5), dpi=100)
# adding more informations on the line
# plt.plot(x,y,label='2x',color='green',linewidth=4,marker='.',markersize=10,markeredgecolor='blue')
# short hand
# format = '[color][marker][line]'
plt.plot(x,y,'g^--',label='2x')

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('First graph', fontdict={'fontname': 'Comic Sans MS','fontsize': 20})

# changing the graph number
plt.xticks([0,1,2,4])
plt.yticks([0,2,4,6,8,10])

# creating curve line
x2 = np.arange(0,4.5,0.5)

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')

# Plot remainder of graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'r--')


# adding infos on the graphs
plt.legend()
plt.show()
```

# Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(5,3), dpi=100)

bars = plt.bar(labels, values)

patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

# plt.savefig('barchart.png', dpi=300)

plt.show()
```
