```python
product_dictionary = []
def get_makro_productName(url,product_type):
    pagination_list = []
    browser.get(url)
    time.sleep(5)
    html = browser.page_source
    soup = bs(html, 'html.parser')
    pagination = soup.find_all(class_='pagination')
    for i in pagination:
        pagination_list.append(i.text)
    if len(pagination_list) > 1:
        print('More than 1')
        for res in pagination_list:
            print(res)
            browser.get(url+'?page='+res+'&sorting=SORTING_LAST_UPDATE&all=true')
            time.sleep(20)
            html = browser.page_source
            soup_pagination = bs(html, 'html.parser')
            print(url+'?page='+res+'&sorting=SORTING_LAST_UPDATE&all=true')
            product_info =  soup_pagination.find_all(class_='ProductCard__CardProduct-sc-1lb6pe3-0 ktTHor')
            for each_product in product_info:
                product_name = each_product.find(class_='ProductCard__TitleMaxLines-sc-1lb6pe3-2 kAKAhd pt-2').text
                product_price_usd = each_product.find(class_='ProductCard__TextPrice-sc-1lb6pe3-1 jcvzXx m-0 mt-lg-1 font-16 color-red weight-bold d-flex align-items-end').text
                product_price_kh = each_product.find(class_='d-flex justify-content-between align-items-center').find('div',class_='m-0 mt-lg-1 font-11 d-flex align-items-end').text
                product_dictionary.append({'product_name': product_name,'typeOf': product_type,'price_usd': product_price_usd, 'price_kh': product_price_kh})
    else:
        browser.get(url+'?all=true')
        time.sleep(20)
        html = browser.page_source
        soup_pagination = bs(html, 'html.parser')
        product_info =  soup_pagination.find_all(class_='ProductCard__CardProduct-sc-1lb6pe3-0 ktTHor')
        for each_product in product_info:
            product_name = each_product.find(class_='ProductCard__TitleMaxLines-sc-1lb6pe3-2 kAKAhd pt-2').text
            product_price_usd = each_product.find(class_='ProductCard__TextPrice-sc-1lb6pe3-1 jcvzXx m-0 mt-lg-1 font-16 color-red weight-bold d-flex align-items-end').text
            product_price_kh = each_product.find(class_='d-flex justify-content-between align-items-center').find('div',class_='m-0 mt-lg-1 font-11 d-flex align-items-end').text
            product_dictionary.append({'product_name': product_name,'typeOf': product_type,'price_usd': product_price_usd, 'price_kh': product_price_kh})
```

# Calling all the items in each page to store in dictionary.

```python
for product in catName:
    get_makro_productName(product,product.split('/')[5])
```

# Converting dict to JSON file.

```python
def save_data(title,data):
    with open(title, "w",encoding='utf-8') as outfile:
        json.dump(data, outfile, ensure_ascii=False, indent=3)
```

```python
def load_data(title):
    with open(title,encoding='utf-8') as f:
        return json.load(f)
```

```python
macro_store_pirce = "./data/macro_store_pirce.json"
save_data(macro_store_pirce,product_dictionary)
macro_store_pirce_list = load_data(macro_store_pirce)
```
