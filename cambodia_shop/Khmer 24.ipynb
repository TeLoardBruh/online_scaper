## Starting scraping process

### Get all category into an array

```python
category_url = []
url = 'https://www.khmer24.com/'

headers = {'User-Agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/105.0.0.0 Mobile Safari/537.36'}

response = requests.get(url, headers=headers)
data = bs(response.content, 'html.parser')

get_ul_parents = data.find(class_="list-inline")

a_href_links = get_ul_parents.find_all('a', href=True)

for a_href_link in a_href_links:
    category_url.append(a_href_link.get('href'))
    
category_url
```
