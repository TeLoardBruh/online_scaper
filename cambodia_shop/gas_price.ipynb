```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Reading data from gas_csv

```python
gas = pd.read_csv('./data/gas_prices.csv')
gas
```

# Plotting data

```python
plt.figure(figsize=(8,5), dpi=250)
plt.title('Gas prices overtime in USD')


# plt.plot(gas['Year'], gas['USA'], 'b.-' ,label='USA')
# plt.plot(gas['Year'], gas['Australia'], 'g.-', label='Australia')
# plt.plot(gas['Year'], gas['France'], 'r.-', label='France')
# plt.plot(gas['Year'], gas['Canada'], 'y.-', label='Canada')

for country in gas:
    if country != 'Year':
        plt.plot(gas['Year'],gas[country],marker='.',label=country)
plt.legend()

plt.xlabel('Year')
plt.ylabel('USD Dollars')
plt.xticks(gas['Year'][::3],rotation = 60, ha="right")
plt.show()
```
